```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

```python
team_stats_df = pd.read_csv('data/team_analysis.csv')
```

```python
team_stats_corr = team_stats_df.corr()
```

```python
fig = plt.figure()
plt.subplots(figsize=(25,25))
seaborn_map = sns.heatmap(team_stats_corr,cmap='seismic')
plt.savefig('Figures/large_heatmap.png',dpi=400)
plt.show()
```

```python
team_stats_cleaned = team_stats_df.drop(columns={'Games','Minutes Played','Field Goals','Field Goal Attempts','3 Point Attempts','3 Point Goals','2 Point Attempts','2 Point Goals','Free Throw Attempts','Free Throws','Total Rebounds','Assists','Steals', 'Blocks', 'Turnovers','Losses','Personal Fouls'})
```

```python
y = team_stats_cleaned.corr()
```

```python
fig = plt.figure()
plt.subplots(figsize=(15,15))
sns.set(font_scale=2)
easy_seaborn_map = sns.heatmap(y, annot=True, cmap='seismic',linewidths=0.25)
plt.tight_layout()
plt.savefig('Figures/readable_heatmap.png',dpi=400)
plt.show()
```

```python
plt.rcParams.update(plt.rcParamsDefault)
```

```python
fig = plt.figure()
a = team_stats_cleaned['Wins']
b = team_stats_cleaned['Points']
correlation = st.pearsonr(a,b)
team_stats_cleaned.plot.scatter(x='Wins',y='Points',edgecolors='k',s=40)
plt.title('Wins Vs Points Scored Per Game For Each Team')
plt.ylabel('Point per Game')
plt.text(40,103,f'Correlation: {round(correlation[0],2)}',fontsize=15,c='b')
plt.savefig('Figures/PointsvsWins.png',dpi=400)
plt.show()
```
